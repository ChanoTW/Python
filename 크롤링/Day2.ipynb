```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20220106'
res = req.get(url)
soup = bs(res.text)
```

```python
title = soup.select('div.tit5>a')
```

# 한달동안의 영화데이터 수집

```python
#1월달 데이터
20220101-20220106
```

```python
title_list = []
score_list = []
rank_list = []

for i in range(20220101,20220107):
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date='+str(i)
    res = req.get(url)
    soup = bs(res.text)
    title = soup.select('div.tit5>a')
    score = soup.select('.point')

    for i in range(len(title)):
        title_list.append(title[i].text)
        score_list.append(score[i].text)
        rank_list.append(i+1)
```

```python
len(title_list)
```

```python
dic = {'영화명':title_list, '평점': score_list, '순위':rank_list}
df = pd.DataFrame(dic)
df.set_index('순위', inplace=True)
df
```

# Pandas활용하여 날짜 생성

```python
date = pd.date_range(start = "2021-01-01", end = "2021-03-31")
date
```

```python
# 문자형 포맷팅으로 필요없는 내용 지우기
# strftime(%년%월%일)
days = date.strftime('%Y%m%d')
```

```python
from tqdm import tqdm_notebook
```

```python
title_list = []
score_list = []
rank_list = []
day_list = []

for day in tqdm_notebook(days):
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date='+day
    res = req.get(url)
    soup = bs(res.text)
    
    title = soup.select('div.tit5>a')
    score = soup.select('.point')
    

    for i in range(len(title)):
        title_list.append(title[i].text)
        score_list.append(score[i].text)
        rank_list.append(i+1)
        day_list.append(day)
```

```python
dic = {'영화명':title_list, '평점': score_list, '순위':rank_list, '날짜':day_list}
df = pd.DataFrame(dic)
df.set_index('순위', inplace=True)
df
```
