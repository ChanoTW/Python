```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
import time
```

```python
driver = wb.Chrome()
time.sleep(0.5)
driver.get("https://www.youtube.com/channel/UC8CIM3d3zDYMk-3T5aAz0yw/videos")
time.sleep(1)
body = driver.find_element_by_css_selector("body")

for i in range(10):
    body.send_keys(Keys.END)
    time.sleep(1)
    
title = driver.find_elements_by_id("video-title")
view = driver.find_elements_by_css_selector("#metadata-line > span:nth-child(1)")
for i in title:
    print(i.text)
for i in view:
    print(i.text)
```

```python
import pandas as pd
```

```python
#창을 안보이게 시작하는 옵션
options = wb.ChromeOptions()
options.add_argument('headless')
options.add_argument('window-size=1920x1080')
options.add_argument("disable-gpu")
```

```python
driver = wb.Chrome("C:/Users/smhrd/크롤링/chromedriver.exe", options = options)
time.sleep(0.5)
driver.get("http://corners.gmarket.co.kr/Bestsellers")
time.sleep(0.5)

title_list = []
price_list = []

for i in range(3):
    img = driver.find_elements_by_css_selector("img.lazy")
    img[i].click()
    time.sleep(1)
    title = driver.find_element_by_css_selector('.itemtit')
    price = driver.find_element_by_css_selector('.price_real')
    title_list.append(title.text)
    price_list.append(price.text)
    driver.back()
    time.sleep(1)
data = {'상품명':title_list,'가격':price_list}
gmarket = pd.DataFrame(data)
gmarket
gmarket.to_csv("지마켓.csv", encoding="euc-kr")
```
